Aqui buscamos a "location" que mais aparece na base de dados.

E calculamos com o "count", a quantidade de vezes que foi listada.

```python
# Encontra a location que mais aparece
location_mais_frequente = df['location'].value_counts().idxmax()

# Conta quantas vezes a location mais frequente aparece
contagem_location = df['location'].value_counts().max()

print(f"O local que mais aparece na base de dados é: {location_mais_frequente}")
print(f"Ele apareceu {contagem_location} vezes.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Encontra a location que mais aparece
location_mais_frequente = df['location'].value_counts().idxmax()

# Conta quantas vezes a location mais frequente aparece
contagem_location = df['location'].value_counts().max()

# Extrai o mês da coluna 'date' para cada linha onde a location é a mais frequente
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
meses_location_frequente = df.loc[df['location'] == location_mais_frequente, 'month']

# Encontra o mês que mais aparece para a location mais frequente
mes_mais_frequente = meses_location_frequente.value_counts().idxmax()

# Conta quantas vezes o mês mais frequente aparece para a location mais frequente
contagem_mes = meses_location_frequente.value_counts().max()

print(f"O local que mais aparece na base de dados é: {location_mais_frequente}")
print(f"Ele apareceu {contagem_location} vezes.")
print(f"O mês com mais menções para {location_mais_frequente} é: {mes_mais_frequente}")
print(f"Ele apareceu {contagem_mes} vezes.")

# Gráfico
plt.figure(figsize=(12, 6))
sns.countplot(x='month', data=df[df['location'] == location_mais_frequente], palette='viridis')
plt.title(f'Frequência de Menções à Location "{location_mais_frequente}" por Mês', fontsize=16)
plt.xlabel('Mês', fontsize=14)
plt.ylabel('Contagem', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adicionar anotações com os valores de contagem acima das barras
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                     textcoords='offset points')


plt.tight_layout()
plt.show()
```

Através do gráfico de pizza, apresentamos as 10 localizações que mais registraram terremotos no mundo.

```python
# Encontra as 10 localizações com mais terremotos
top_10_locations = df['location'].value_counts().nlargest(10)

# Cria o gráfico de pizza com detalhes
plt.figure(figsize=(10, 8))  # Define o tamanho da figura
plt.pie(top_10_locations, labels=top_10_locations.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"), wedgeprops={'linewidth': 1, 'edgecolor': 'white'})  # Personaliza cores e bordas
plt.title('Top 10 Locais com Maior Número de Terremotos', fontsize=16)  # Define o título
plt.axis('equal')  # Garante que o gráfico seja um círculo
plt.legend(title="Localizações", loc="best", bbox_to_anchor=(1, 0, 0.5, 1))  # Adiciona legenda fora do gráfico

plt.show()
```

Nessa pesquisa, buscamos a média de magnitude dos terremotos ocasionados nessas 10 localizações mais comuns.

```python
import plotly.express as px

# Agrupar por location e calcular a média da magnitude
media_magnitude_por_location = df.groupby('location')['magnitude'].mean()

# Filtrar as 10 locations mais frequentes
top_10_locations_media = media_magnitude_por_location[top_10_locations.index]

# Criar o gráfico de barras com Plotly
fig = px.bar(top_10_locations_media, x=top_10_locations_media.index, y=top_10_locations_media.values,
             labels={'x': 'Localização', 'y': 'Média de Magnitude'},
             title='Média de Magnitude dos Terremotos por Localização (Top 10)')
fig.show()
```

```python
# Converter a coluna 'date' para o tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Extrair o mês da coluna 'date'
df['mes'] = df['date'].dt.month

# Filtrar as 10 localizações mais frequentes
top_10_locations = df['location'].value_counts()[:10].index

# Criar um novo DataFrame apenas com as 10 localizações mais frequentes
df_top_10 = df[df['location'].isin(top_10_locations)]

# Contar a frequência dos meses em que ocorreram terremotos nas 10 localizações mais frequentes
contagem_meses = df_top_10['mes'].value_counts()

# Criar um gráfico de barras com Plotly para visualizar a frequência dos meses
fig = px.bar(contagem_meses, x=contagem_meses.index, y=contagem_meses.values,
             labels={'x': 'Mês', 'y': 'Frequência de Terremotos'},
             title='Frequência de Terremotos por Mês (Top 10 Localizações)')

# Configurar o eixo x para exibir os nomes dos meses em português
fig.update_xaxes(
    tickmode='array',
    tickvals=list(range(1, 13)),
    ticktext=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
              'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
)

fig.show()
```

```python
# 10 localizações com mais terremotos
top_10_locations = df['location'].value_counts().nlargest(10).index

# Filtra o DataFrame para incluir apenas as 10 localizações mais frequentes
df_top_10 = df[df['location'].isin(top_10_locations)]

# Filtra o DataFrame para incluir apenas as linhas onde tsunami = 1
df_top_10_tsunami = df_top_10[df_top_10['tsunami'] == 1]

# Conta quantas vezes cada localização aparece com tsunami = 1
contagem_tsunami_por_location = df_top_10_tsunami['location'].value_counts()

# Cria o gráfico de pizza com detalhes
plt.figure(figsize=(10, 8))  # Define o tamanho da figura
plt.pie(contagem_tsunami_por_location, labels=contagem_tsunami_por_location.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"), wedgeprops={'linewidth': 1, 'edgecolor': 'white'})  # Personaliza cores e bordas
plt.title('Localizações com Tsunami (entre os 10 locais com mais frequência de terremotos)', fontsize=16)  # Define o título
plt.axis('equal')  # Garante que o gráfico seja um círculo
plt.legend(title="Localizações", loc="best", bbox_to_anchor=(1, 0, 0.5, 1))  # Adiciona legenda fora do gráfico

plt.show()
```
